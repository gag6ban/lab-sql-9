```python
# connection_string helps to find where is everything
connection_string = 'mysql+pymysql://root:'+password+'@localhost/bank'
# either localhost or ip address
engine = create_engine(connection_string)
```

```python
# Q1: How many distinct (different) actors' last names are there?
# Q2: Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
# Q3: Get all films with ARMAGEDDON in the title.
# Q4: Get 10 the longest films.
# Q5: How many films include Behind the Scenes content?
# Q6: Which kind of movies (rating) have a mean duration of more than two hours?
# Q7: Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
```

```python
# Q1: How many distinct (different) actors' last names are there?

query1 = 'SELECT COUNT(DISTINCT last_name) FROM sakila.actor'
data = pd.read_sql_query(query1, engine)
data.head()
```

```python
# Q2: Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

query2= '''SELECT rental_id, rental_date, CASE WHEN WEEKDAY(rental_date) IN (5,6) THEN 'Weekend' ELSE 'Workday' END AS 'Weekday' 
FROM sakila.rental'''

data = pd.read_sql_query(query2, engine)
data.head()
```

```python
# Q3: Get all films with ARMAGEDDON in the title.

# query3 = 'SELECT title FROM sakila.film WHERE title = '%ARMAGEDDON%''

data = pd.read_sql_query('SELECT title FROM sakila.film', engine)
data[data['title'].str.contains("ARMAGEDDON")]
data.head()
```

```python
# Q4: Get 10 the longest films.

query4= 'SELECT * FROM sakila.film ORDER BY (length) DESC LIMIT 10'

data = pd.read_sql_query(query4, engine)
data.head()
```

```python
# Q5: How many films include Behind the Scenes content?

query5= 'SELECT COUNT(title) FROM sakila.film WHERE special_features LIKE '%Behind the Scenes%';'
data = pd.read_sql_query(query5, engine)
data.head()
```

```python
# Q6: Which kind of movies (rating) have a mean duration of more than two hours?

query6 = 'SELECT DISTINCT rating, COUNT(*), AVG(length)/60 as avg FROM sakila.film GROUP BY rating HAVING avg > 2;'

data = pd.read_sql_query(query6, engine)
data.head()
```

```python
# Q7: Rank films by length (filter out the rows that have nulls or 0s in length column). 
# In your output, only select the columns title, length, and the rank.

# query7 = 'SELECT title, length FROM sakila.film where length !=0 AND length is not null order by length DESC'
# data = pd.read_sql_query(query7, engine)
# data.head()


query7: '''SELECT DENSE_RANK() OVER(ORDER BY length DESC) as 'Rank', title, length FROM film WHERE length != 0 and length IS NOT NULL'''
data = pd.read_sql_query(query7, engine)
data.head()    
```
